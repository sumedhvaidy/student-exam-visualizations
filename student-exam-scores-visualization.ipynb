```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
#Math scores
#Gender
bins = ['0-10', '11-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-90', '91-100']
females = []
males = []
lowerBound = 0
upperBound = 10
while(upperBound <= 100):
    females.append(df['gender'][(df['gender'] == 'female') & (df['math score'] >= lowerBound) & (df['math score'] <= upperBound)].count())
    males.append(df['gender'][(df['gender'] == 'male') & (df['math score'] >= lowerBound) & (df['math score'] <= upperBound)].count())
    if(lowerBound == 0):
        lowerBound += 11
        upperBound += 10
    else:
        lowerBound += 10
        upperBound += 10
females = np.array(females)
males = np.array(males)
ind = [bin for bin in bins]

plt.bar(ind, males, width=0.6, label='males', color='gold', bottom=females)
plt.bar(ind, females, width=0.6, label='females', color='silver')
 
plt.xticks(ind, bins)
plt.ylabel("Frequency")
plt.xlabel("Math Exam Scores")
plt.legend(loc="upper right")
plt.title("Math Exam Scores based off Gender")
plt.show()
```

```python
#Math scores
#Gender
bins = ['0-10', '11-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-90', '91-100']
groupA = []
groupB = []
groupC = []
groupD = []
groupE = []
lowerBound = 0
upperBound = 10
while(upperBound <= 100):
    groupA.append(df['race/ethnicity'][(df['race/ethnicity'] == 'group A') & (df['math score'] >= lowerBound) & (df['math score'] <= upperBound)].count())
    groupB.append(df['race/ethnicity'][(df['race/ethnicity'] == 'group B') & (df['math score'] >= lowerBound) & (df['math score'] <= upperBound)].count())
    groupC.append(df['race/ethnicity'][(df['race/ethnicity'] == 'group C') & (df['math score'] >= lowerBound) & (df['math score'] <= upperBound)].count())
    groupD.append(df['race/ethnicity'][(df['race/ethnicity'] == 'group D') & (df['math score'] >= lowerBound) & (df['math score'] <= upperBound)].count())
    groupE.append(df['race/ethnicity'][(df['race/ethnicity'] == 'group E') & (df['math score'] >= lowerBound) & (df['math score'] <= upperBound)].count())
    if(lowerBound == 0):
        lowerBound += 11
        upperBound += 10
    else:
        lowerBound += 10
        upperBound += 10
groupA = np.array(groupA)
groupB = np.array(groupB)
groupC = np.array(groupC)
groupD = np.array(groupD)
groupE = np.array(groupE)
ind = [bin for bin in bins]
plt.bar(ind, groupE, width=0.6, label='Group E', color='green', bottom=groupD+groupC+groupB+groupA)
plt.bar(ind, groupD, width=0.6, label='Group D', color='blue', bottom=groupC+groupB+groupA)
plt.bar(ind, groupC, width=0.6, label='Group C', color='red', bottom=groupB+groupA)
plt.bar(ind, groupB, width=0.6, label='Group B', color='gold', bottom=groupA)
plt.bar(ind, groupA, width=0.6, label='Group A', color='silver')
 
plt.xticks(ind, bins)
plt.ylabel("Frequency")
plt.xlabel("Math Exam Scores")
plt.legend(loc="upper right")
plt.title("Math Exam Scores based off Race / Ethnicity")
plt.show()
```
